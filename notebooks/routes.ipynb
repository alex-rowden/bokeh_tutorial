```python
FILE = "../data/T_T100D_MARKET_US_CARRIER_ONLY_20220506_205137.zip"

import pandas as pd

from bokeh.io import output_notebook
output_notebook() 

domestic_carriers_df = pd.read_csv(
    FILE,
    compression="zip",
    usecols=range(22),
    dtype={
        "PASSENGERS": "int64",
        "FREIGHT": "int64",
        "MAIL": "int64",
        "DISTANCE": "int64",
        },
    )
domestic_carriers_df.columns = map(str.lower, domestic_carriers_df.columns)
domestic_carriers_df.head()
```

```python
# find routes with the most passengers per kilometer
# Or: longest routes (10 longest routes, 100 longest routes, etc)
```

```python
domestic_carriers_df["route"] = domestic_carriers_df["origin"] + "-" + domestic_carriers_df["dest"]
```

```python
# df of largest routes by passengers
routes_df = domestic_carriers_df.groupby("route").agg({"passengers": "sum"}).sort_values("passengers", ascending=False).iloc[:100]
routes_df
```
